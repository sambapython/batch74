```python
data = [{'id': 1, 'name': 'jayaram', 'age': 4}, 
        {'id': 2, 'name': 'vedhanvesh', 'age': 3}, 
        {'id': 3, 'name': 'sai', 'age': 32}, 
        {'id': 4, 'name': 'samba', 'age': 33}]
f=open("data.txt","w")
for row in data:
    f.write(",".join(row.values()))
f.close()
```

```python
data = [{'id': 1, 'name': 'jayaram', 'age': 4}, 
        {'id': 2, 'name': 'vedhanvesh', 'age': 3}, 
        {'id': 3, 'name': 'sai', 'age': 32}, 
        {'id': 4, 'name': 'samba', 'age': 33}]
f=open("data.txt","w")
for row in data:
    row = ",".join(list(map(str,row.values())))
    row=row+"\n"
    f.write(row)
f.close()
```

```python
f=open("data.txt")
delimerter="."
row = ""
while True:
    letter = f.read(1)
    if letter == "":
        print(row)
        break
    elif letter==delimerter:
        print(row)
        row=""
    else:
        row = row+letter
```
